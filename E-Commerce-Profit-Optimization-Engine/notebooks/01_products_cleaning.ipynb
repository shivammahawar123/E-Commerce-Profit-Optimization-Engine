```python
import pandas as pd 
```

```python
df = pd.read_csv("../data/walmart_products.csv")
```

```python
df = df[[
    "Product Name",
    "Brand",
    "Category",
    "List Price",
    "Sale Price",
    "Available"
]]
```

```python
df
```

```python
df.columns = [
    "product_name",
    "brand",
    "category",
    "list_price",
    "sale_price",
    "available"
]
```

```python
df.isnull().sum()
```

```python
df["brand"] = df["brand"].fillna("Unknown")

df = df.dropna(subset=["category"])
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df["list_price"] = pd.to_numeric(df["list_price"])

df["sale_price"] = pd.to_numeric(df["sale_price"])
```

```python
df[["list_price", "sale_price"]].describe()
```

```python
df = df[
    (df["list_price"] > 0) &
    (df["sale_price"] > 0)
]
```

```python
df[["list_price", "sale_price"]].describe()
```

```python
df["discount_pct"] = (
    (df["list_price"] - df["sale_price"])
    / df["list_price"]
) * 100

df["discount_pct"] = df["discount_pct"].round(2)
```

```python
df[["list_price", "sale_price", "discount_pct"]].head()
```

```python
df.insert(0, "product_id", range(1, len(df) + 1))
```

```python
df["main_category"] = df["category"].str.split("|").str[0].str.strip()

df["subcategory"] = df["category"].str.split("|").str[1].str.strip()
```

```python
df[["category", "main_category", "subcategory"]].head()
```

```python
df = df[[
    "product_id",
    "product_name",
    "brand",
    "main_category",
    "subcategory",
    "list_price",
    "sale_price",
    "discount_pct",
    "available"
]]
```

```python
df
```

```python
df.to_csv(
    "../data/products.csv",
    index=False
)
```
