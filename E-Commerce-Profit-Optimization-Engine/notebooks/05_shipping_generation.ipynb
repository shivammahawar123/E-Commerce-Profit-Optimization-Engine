```python
import pandas as pd
import random
```

```python
orders_df = pd.read_csv(
    "../data/orders.csv"
)
```

```python
shipping = []
```

```python
shipping_providers = [
    "FedEx",
    "UPS",
    "USPS",
    "DHL"
]

shipping_regions = [
    "West",
    "East",
    "South",
    "Midwest"
]

shipping_statuses = [
    "Delivered",
    "In Transit",
    "Delayed"
]

for i in range(len(orders_df)):

    order = orders_df.iloc[i]

    shipping.append({
        "shipping_id": i + 1,
        "order_id": order["order_id"],
        "shipping_provider": random.choice(
            shipping_providers
        ),
        "shipping_region": random.choice(
            shipping_regions
        ),
        "delivery_days": random.randint(1, 10),
        "shipping_status": random.choice(
            shipping_statuses
        )
    })
```

```python
shipping_df = pd.DataFrame(shipping)
```
