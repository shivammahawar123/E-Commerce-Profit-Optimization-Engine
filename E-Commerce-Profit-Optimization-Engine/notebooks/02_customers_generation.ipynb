```python
import pandas as pd
import random
from faker import Faker
```

```python
fake = Faker()
```

```python
num_customers = 10000
```

```python
customers = []
```

```python
city_state_pairs = [
    ("New York", "New York"),
    ("Los Angeles", "California"),
    ("Chicago", "Illinois"),
    ("Houston", "Texas"),
    ("Phoenix", "Arizona"),
    ("Philadelphia", "Pennsylvania"),
    ("San Antonio", "Texas"),
    ("San Diego", "California"),
    ("Dallas", "Texas"),
    ("San Jose", "California")
]

customers = []

for i in range(1, num_customers + 1):

    gender = random.choice(["Male", "Female"])

    if gender == "Male":
        customer_name = fake.name_male()
    else:
        customer_name = fake.name_female()

    city, state = random.choice(city_state_pairs)

    customers.append({
        "customer_id": i,
        "customer_name": customer_name,
        "gender": gender,
        "city": city,
        "state": state,
        "signup_date": fake.date_between(
            start_date='-3y',
            end_date='today'
        ),
        "customer_segment": random.choice([
            "Regular",
            "Premium",
            "VIP"
        ])
    })
```

```python
customers_df = pd.DataFrame(customers)
```
